```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\rushi\\Desktop\\DPL\\predict2023_1.csv")
df.head(10)
```

```python
df=df.drop('Notes',axis=1)
df.head(10)
```

```python
df=df.drop('Unnamed: 0',axis=1)
df=df.drop('Match Report',axis=1)
df.head(10)
```

```python
#This plot shows the time series analysis between goals for and time which can be used to provide
# valuable insights into historical performance, which can be leveraged for predictions, strategic planning, and a deeper
# understanding of the dynamics of Premier League matches.
df['Date'] = pd.to_datetime(df['Date'])

# Sort the DataFrame by 'Date' for a chronological order
df = df.sort_values(by='Date')

# Create a time series plot
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['GF'], marker='o', linestyle='-')
plt.title('Time Series Analysis: Goals For (GF) Over Time')
plt.xlabel('Date')
plt.ylabel('Goals For (GF)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for readability

# Show the plot
plt.show()
```

```python
team_goals = df.groupby('Team')['GF'].sum().reset_index()

# Sort the data by total goals scored in descending order
team_goals = team_goals.sort_values(by='GF', ascending=False)

# Create a bar plot to show how much each team has scored
plt.figure(figsize=(12, 6))
sns.barplot(x='GF', y='Team', data=team_goals, palette='colorblind') #colorblind is color palette specially used for presentation.
plt.title('Total Goals Scored by Each Team')
plt.xlabel('Total Goals For (GF)')
plt.ylabel('Team')

# Show the plot
plt.show()
```

```python
# In this plot we come to know that results are also affected by the venue which can be a point for predicting.
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.countplot(x='Venue', hue='Result', data=df, hue_order=['W', 'L'], palette='Set2')
plt.title('Venue vs. Results for Each Team')
plt.xlabel('Venue')
plt.ylabel('Count')

# Show the plot
plt.legend(title='Result')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot for Goals For (GF)
sns.barplot(x='GF', y='Team', data=df, palette='colorblind', ax=axes[0])
axes[0].set_title('Goals For (GF)')
axes[0].set_xlabel('Goals For (GF)')
axes[0].set_ylabel('Team')

# Plot for Goals Against (GA)
sns.barplot(x='GA', y='Team', data=df, palette='colorblind', ax=axes[1])
axes[1].set_title('Goals Against (GA)')
axes[1].set_xlabel('Goals Against (GA)')
axes[1].set_ylabel('')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(x='Passes', y='Shots', data=df1, hue='Team', palette='Set1')

# Add labels for each data point
for i in range(len(df1)):
    plt.text(df1['Passes'][i] + 0.2, df1['Shots'][i], df1['Team'][i], fontsize=8)

plt.title('Teamwise Passes vs Shots per game')
plt.xlabel('Passes')
plt.ylabel('Shots')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
